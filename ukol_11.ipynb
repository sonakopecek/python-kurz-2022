Stáhni si soubor platy_2021_02.csv s informacemi o platech v softwarové firmě.

Načti si tato data do tabulky a vytvoř histogram. Nastav vhodně hranice skupin histogramu (parametr bins), aby byl graf přehledný a snadno interpretovatelný

```python
import pandas as pd
import matplotlib.pyplot as plt
import pytemperature
platy = pd.read_csv("platy_2021_02.csv")
platy = platy.drop(platy.columns[0], axis=1)
platy
```

```python
platy.hist()
```

```python
platy.hist(bins=11)
plt.show()
```

```python
platy.hist(bins=12, rwidth = 0.9, facecolor = "#358600")
plt.title("Platy za únor 2021\n\n", fontweight ="bold")
plt.xlabel('výše platu v Kč')
plt.ylabel('zaměstnanci')
```

Vrať se k práci se souborem temperature.csv, který obsahuje informace o průměrné teplotě v různých městech v listopadu 2017. 
- Vytvoř tabulku, která bude obsahovat údaje o teplotě za města Helsinki, Miami Beach a Tokyo.
- Vytvoř krabicový graf a porovnej rozsah teplot v těchto městech.

```python
teploty = pd.read_csv("temperature.csv")
teploty["AvgTemperatureCelsia"] = pytemperature.f2c(teploty["AvgTemperature"])
teploty.head()
```

```python
teploty_vybrane = teploty[teploty["City"].isin(["Helsinki", "Miami Beach", "Tokyo"])]
teploty_vybrane
```

```python
Helsinki = teploty_vybrane[teploty_vybrane["City"] == "Helsinki"]["AvgTemperatureCelsia"]
Miami_Beach = teploty_vybrane[teploty_vybrane["City"] == "Miami Beach"]["AvgTemperatureCelsia"]
Tokyo = teploty_vybrane[teploty_vybrane["City"] == "Tokyo"]["AvgTemperatureCelsia"]

plt.boxplot([Helsinki, Miami_Beach, Tokyo], labels=['Helsinki', 'Miami Beach', 'Tokyo'])

plt.title("Porovnání rozsahu teplot v Helsinkách, Miami Beach a Tokyu", fontweight ="bold")
plt.xlabel("Město")
plt.ylabel("Teplota (°C)")
plt.show()
```
